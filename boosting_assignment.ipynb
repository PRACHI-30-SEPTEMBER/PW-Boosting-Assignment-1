# Boosting in Machine Learning - Assignment (Q1–Q10)

Questions and Answers

## Q1. What is boosting in machine learning?

**Answer:** Boosting is an ensemble learning technique that combines multiple weak learners (usually decision trees) to form a strong learner. It works by training models sequentially, where each new model focuses on correcting the errors made by previous models.

## Q2. Advantages and limitations of boosting techniques?

**Answer:** Advantages:
- High accuracy
- Handles bias well
- Works with different loss functions

Limitations:
- Sensitive to noise
- Can overfit if not tuned properly
- Computationally expensive

## Q3. Explain how boosting works.

**Answer:** Boosting works by training models sequentially. Each model gives more importance (weight) to incorrectly predicted samples so that future models focus more on difficult cases.

## Q4. Different types of boosting algorithms?

**Answer:** Common types:
- AdaBoost
- Gradient Boosting
- XGBoost
- LightGBM
- CatBoost

## Q5. Common parameters in boosting algorithms?

**Answer:** Some common parameters include:
- n_estimators
- learning_rate
- max_depth
- subsample
- min_samples_split

## Q6. How do boosting algorithms combine weak learners?

**Answer:** They combine weak learners in a sequential manner where each learner improves the previous one. Final prediction is usually a weighted sum or vote of all models.

## Q7. Explain AdaBoost algorithm.

**Answer:** AdaBoost assigns weights to all training samples. Initially all weights are equal. After each iteration, misclassified samples get higher weights so that next model focuses more on them.

## Q8. Loss function used in AdaBoost?

**Answer:** AdaBoost typically uses exponential loss function.

## Q9. How does AdaBoost update weights?

**Answer:** Weights of misclassified samples are increased, and weights of correctly classified samples are decreased, making the model focus more on difficult cases.

## Q10. Effect of increasing number of estimators?

**Answer:** Increasing estimators generally improves performance up to a point, but too many estimators can lead to overfitting and increased computation time.
